```python
from pickle import NONE
class Mahasiswa:
    def __init__(self, nama, npm, email, jurusan, angkatan,universitas):
        self._nama = nama
        self._npm = npm
        self._email = email
        self._jurusan = jurusan
        self._angkatan = angkatan
        self._universitas = universitas

    def data_mahasiswa(self):
        return f"nama: {self._nama}\nnpm: {self._npm}\nEmail: {self._email}\njurusan: {self._jurusan}\nangkatan: {self._angkatan}\nUniversitas :{self._universitas}\n--------------------------------------"

class database_mahasiswa(Mahasiswa):
    def __init__(self):
        self._data = []

    def __str__(self):
      #jaga jaga jika database kosong
        if not self._data:
            return "data mahasiswa masih kosong!"

        data_mahasiswaa = "\n".join(Mahasiswa.data_mahasiswa() for Mahasiswa in self._data)
        return f"Data Mahasiswa Sebagai Berikut:\n{data_mahasiswaa}"

    def tambah_mahasiswa(self, mahasiswa):
        self._data.append(mahasiswa)

    def hapus_mahasiswa(self, mahasiswa):
        try:
            self._data.remove(mahasiswa)
        except ValueError:
            print("tidak ada data mahasiswa ini")

    def cari_mahasiswa(self, ma):
        for mahasiswa in self._data:
            if mahasiswa._nama == ma:
                return mahasiswa.data_mahasiswa()
        return None

    def cari_NPM(self, npm):
        for NPM in self._data:
            if NPM._npm == npm:
                return NPM.data_mahasiswa()
        return None

    def cari_Jurusan(self, Jurusan):
        for Jurusan in self._data:
            if Jurusan.get_Jurusan() == Jurusan:
                return Jurusan
        return None


if __name__ == "__main__":
    database = database_mahasiswa()

    NPM_str = [str(i) for i in NPM]
    NPM_2_str = [str(i) for i in NPM_2]
    NPM_3_str = [str(i) for i in NPM_3]

    Nama.extend(Nama_2)
    Nama.extend(Nama_3)
    NPM_str.extend(NPM_2_str)
    NPM_str.extend(NPM_3_str)
    Email.extend(Email_2)
    Email.extend(Email_3)
    Jurusan.extend(Jurusan.copy())
    Jurusan.extend(Jurusan.copy())


    datamahasiswa = zip(Nama,NPM_str,Email,Jurusan)
    for j in datamahasiswa:
      data = Mahasiswa(*j,Angkatan,universitas(NPM_str))
      database.tambah_mahasiswa(data)
    print(database.__str__())
```
